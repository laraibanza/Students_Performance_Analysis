```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('StudentsPerformance.csv')
df
```

```python
# checking the shape of df
df.shape
```

```python
# getting the insight of df
df.head()
```

```python
# checking for null values
df.isnull().sum()
```

```python
# checking for duplicated values
df.duplicated().sum()
```

```python
# adding the column average score based on 3 columns
df['average score']= (df['math score']+df['reading score']+df['writing score'])/3
df
```

```python
# adding the result column based on average score column created recently
df['result'] = df['average score'].apply(lambda x : 'Pass' if x>=60 else 'Fail')
df
```

```python
df.tail()
```

```python
# Descriptive Statistics
df.describe()
```

```python
df.gender.value_counts()
```

```python
df['test preparation course'].value_counts()
```

```python
# Group wise Analysis
df.groupby('gender')['average score'].mean()
```

```python
# Result based on gender
df.groupby('gender')['result'].value_counts()
```

```python
# parents education impact on results of studenrs
df.groupby('parental level of education')['result'].value_counts()
```

```python
# Pass rate by test preparation course
df.groupby('test preparation course')['result'].value_counts()
```

```python
# test completion based on gender
df.groupby('gender')['test preparation course'].value_counts()
```

```python
# gender belong to different groups
df.groupby('gender')['race/ethnicity'].value_counts()
```

```python
# Distribution of Scores
sns.histplot(df['average score'], kde=True)
plt.title("Distribution of Average Scores")
plt.show()
```

```python
# Scores by Gender
sns.barplot(x='gender', y='average score', data=df)
plt.title("Average Scores by Gender")
plt.show()
```

```python
# Test Preparation Impact
sns.barplot(x='test preparation course', y='average score', data=df)
plt.title("Effect of Test Prep on Scores")
plt.show()
```
